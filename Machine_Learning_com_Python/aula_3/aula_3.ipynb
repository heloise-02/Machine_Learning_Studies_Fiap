```python
df_ultima_decada = df[df['release_date'].between(pd.to_datetime('2011'), pd.to_datetime('2020'))]
```

```python
def classifica_musica_ao_vivo(df):
    if df["liveness"] >= 0.8:
        return True
    else:
        return False
```

```python
df['ao_vivo'] =df.apply(classifica_musica_ao_vivo, axis = 1)
```

```python
df.head()
```

```python
df.groupby('ao_vivo')['ao_vivo'].count()
```

```python
df_gravado_em_studio = df[df['ao_vivo'] == False]
df_show_ao_vivo = df[df['ao_vivo'] == True]
```

```python
fig, axes=plt.subplots(figsize=(12,6))
sns.violinplot(x='duracao_em_min', data=df_show_ao_vivo, color='darkgray')
axes.set_title('Gráfico Violin das Musicas ao Vivo')
plt.show()
```

```python
fig, axes=plt.subplots(figsize=(12,6))
sns.violinplot(x='duracao_em_min', data=df_gravado_em_studio, color='darkgray')
sns.violinplot(x='duracao_em_min', data=df_show_ao_vivo, color='blue', alpha=0.5)
axes.set_title('Gráfico Violin das Músicas Gravadas no Studio x Músicas Ao Vivo')
plt.show()
```

```python
df_studio = df_gravado_em_studio.groupby('album')['loudness'].sum()
df_ao_vivo = df_show_ao_vivo.groupby('album')['loudness'].sum()

fig, axes =plt.subplots(1, 2, figsize=(15,4))

sns.histplot(data=df_studio, bins=20, ax=axes[0])
axes[0].set_title('Soma dos barulhos dos albuns no studiio')
axes[0].set_xlabel('Barulho')
axes[0].set_ylabel('Frequência')

sns.histplot(data=df_ao_vivo, bins=20, ax=axes[1])
axes[1].set_title('Soma dos Bbrulhos dos albuns ao vivo')
axes[1].set_xlabel('Barulho')
axes[1].set_ylabel('Frequência')

fig.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.kdeplot(data=df_studio, label='Album de studio', fill=True)
sns.kdeplot(data=df_ao_vivo, label='album ao vivo', fill=True)

plt.title('Distribuição do barulho dos albuns')
plt.xlabel('Barulho')
plt.ylabel('Densidade')
plt.legend()
```

```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro(df_studio)
print('Soma do Barulho dos albuns de studio: ')
print('Estáticas de teste: {:.4f}, valor p: {}'.format(stat, p))

if p > 0.05:
    print('Não há evidência sufuciente para rejeitar a hipótese de normalidade')
else:
    print('A hipótes de normalidade é rejeitada')
```

```python
stat, p = shapiro(df_ao_vivo)
print('Soma do Barulho dos albuns ao vivo: ')
print('Estáticas de teste: {:.4f}, valor p: {}'.format(stat, p))

if p > 0.05:
    print('Não há evidência suficiente para rejeitar a hipótes de normalidade')
else:
    print('A hispótese de normalidade é rejeitada')
```

```python
from scipy.stats import mannwhitneyu
```

```python
stat, p = mannwhitneyu(df_studio.sample(len(df_studio)), df_ao_vivo.sample (len(df_ao_vivo)), alternative='less')

print('Estatícas de teste U: ', stat)
print('Valor p:', p)

alpha = 0.05
if p < alpha:
    print('Diferença estatisticamente significante')
else:
    print('Não há diferença estatisticamente significante')
```

```python
media_por_album = df.groupby('album')['valence'].mean().reset_index()

media_por_album = media_por_album.rename(columns={'valence': 'media_valence'})

media_por_album['sentimento'] = ['positivo' if v >0.6 else 'negativo' for v in media_por_album['media_valence']]

media_por_album.groupby('sentimento')['sentimento'].count()
```

```python
media_por_album.head(2)
```

```python
df_resultado_final = pd.merge(df, media_por_album, on ='album')
df_resultado_final.head(2)
```

```python
matriz_correlacao = df_resultado_final.select_dtypes(include=[float, int]).corr()
correlacao_sentimento = matriz_correlacao['media_valence']
display(correlacao_sentimento)
```

```python
sns.heatmap(correlacao_sentimento.to_frame(), annot=True, cmap = 'coolwarm')
plt.show()
```

```python
sns.scatterplot(x='media_valence', y='danceability', hue='sentimento', data=df_resultado_final, palette='coolwarm')
plt.xlabel('Média de Valência (Valence)')
plt.ylabel('Danceabilidade (Danceability)')
plt.title('Relação entre Valência Média e Danceabilidade com Sentimento')
plt.show()
```

```python
sns.scatterplot(x='media_valence', y='liveness', hue='sentimento', data=df_resultado_final, palette='coolwarm')
plt.xlabel('Média de Valência')
plt.ylabel('Liveness')
plt.title('Relação entre Valência Média e Liveness em Músicas')
plt.show()
```
