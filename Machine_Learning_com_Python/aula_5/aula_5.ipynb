```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
x= imoveis[['Area', 'Suites', 'IA', 'Semruido', 'Vista', 'Andar', 'AV100m', 'DistBM']]
y= imoveis[['Valor']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test, = train_test_split(x,y, test_size=0.2, random_state=101)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
print('Intercepto: ', lr.intercept_)
```

```python
coefficients = pd.concat([pd.DataFrame(x.columns), pd.DataFrame(np.transpose(lr.coef_))], axis= 1)
coefficients
```

```python
fig = plt.figure(figsize=(8,6), dpi=80)
plt.rcParams.update({'font.size': 14})
ax = sb.regplot(x = y_test, y = y_pred)
ax.set(xlabel='y real', ylabel='y predito')
ax = plt.plot(y_test, y_test, '--r')

plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
MAE = mean_squared_error(y_test, y_pred)
MSE = mean_squared_error (y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MAE', MAE)
print('MSE', MSE)
print('r2', r2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dtr = DecisionTreeClassifier(random_state=101, max_depth=10)
model_dtr.fit(x_train, y_train)
```

```python
y_pred_model_dtr = model_dtr.predict(x_test)
```

```python
MAE = mean_absolute_error(y_test, y_pred_model_dtr)
MSE = mean_absolute_error(y_test, y_pred_model_dtr)
r2 = r2_score(y_test, y_pred_model_dtr)
print('MAE', MAE)
print('MSE', MSE)
print('R2', r2)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel='linear')
```

```python
svr.fit(x_train, y_train)
```

```python
y_pred_svr = svr.predict(x_test)
```

```python
MAE = mean_squared_error(y_test, y_pred_svr)
MSE = mean_absolute_error(y_test, y_pred_svr)
r2 = r2_score(y_test, y_pred_svr)
print('MAE', MAE)
print('MSE', MSE)
print('R2', r2)
```
